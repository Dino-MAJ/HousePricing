```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['median_house_value'], axis = 1)
y = data["median_house_value"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
train_data = X_train.join(y_train)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only=True), annot = True, cmap = "YlGnBu")
```

```python
# Normalizing these four features
train_data["total_rooms"] = np.log(train_data["total_rooms"] + 1)
train_data["total_bedrooms"] = np.log(train_data["total_bedrooms"] + 1)
train_data["population"] = np.log(train_data["population"] + 1)
train_data["total_rooms"] = np.log(train_data["households"] + 1)
```

```python
train_data.ocean_proximity.value_counts()
```

```python
#One-Hot encoding for ocean proximity
pd.get_dummies(train_data.ocean_proximity)
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(["ocean_proximity"], axis = 1)
```

```python
# Making new features to train on
train_data["bedroom_ratio"] = train_data["total_bedrooms"]/train_data["total_rooms"]
train_data["household_rooms"] = train_data["total_rooms"]/train_data["households"]
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only=True), annot = True, cmap = "YlGnBu")
```

```python
# Linear Regression Model
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#Need to recategroize the data as we added two new feature variables, but the splot can stay the same

X_train, y_train = train_data.drop(["median_house_value"], axis = 1), train_data["median_house_value"]
X_train_s = scaler.fit_transform(X_train)
reg = LinearRegression()
reg.fit(X_train_s, y_train)
```

```python
# Need to add all the test_data together and make all the changes we did before because
# We split up the data set tp early before making all of our feature changes
test_data = X_test.join(y_test)

test_data["total_rooms"] = np.log(test_data["total_rooms"] + 1)
test_data["total_bedrooms"] = np.log(test_data["total_bedrooms"] + 1)
test_data["population"] = np.log(test_data["population"] + 1)
test_data["total_rooms"] = np.log(test_data["households"] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(["ocean_proximity"], axis = 1)

test_data["bedroom_ratio"] = test_data["total_bedrooms"]/test_data["total_rooms"]
test_data["household_rooms"] = test_data["total_rooms"]/test_data["households"]
```

```python
X_test, y_test = test_data.drop(["median_house_value"], axis=1), test_data["median_house_value"]
```

```python
# Random Forest Mode, will also do hyperparameters tuning
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test, y_test)
```

```python
# Cross validation is we split in into k-folds, and have 10 dif. k-folds, nine are used
# For training while 1 is used for test, and the program runs through all possible 
# Combinations of those hyperparameters and spits out the best result
from sklearn.model_selection import GridSearchCV

forest  = RandomForestRegressor() #Defining a fresh one so we don't have fitted version from above
# Change the features to be a little higher and add new ones to try for better hyperparameters
param_grid = {
    "n_estimators": [100, 300],
    "max_depth": [None, 4],
    "min_samples_split": [2, 4]
}
grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring = "neg_mean_squared_error",
                           return_train_score=True)
grid_search.fit(X_train_s, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(X_test_s, y_test)
```
